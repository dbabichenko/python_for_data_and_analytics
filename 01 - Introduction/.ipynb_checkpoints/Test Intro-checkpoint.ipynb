```python
source_file = open('federalist_papers.txt')
all_text = source_file.read()


#print(all_text)

word_list = all_text.split(" ")
#print(word_list)

freq_while = 0
freq_whilst = 0

for word in word_list:
    if word == 'while':
        freq_while = freq_while + 1
    if word == 'whilst':
        freq_whilst = freq_whilst + 1
        
print(freq_while)
print(freq_whilst)
```
