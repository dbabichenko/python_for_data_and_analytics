```python
import numpy as np
import matplotlib.pyplot as pp
```

```python
# Instructs matplotlib to keep plots in the notebook itself
%matplotlib inline


# Create a sample array
x = np.linspace(0, 10, 40)
x
```

```python
sinx = np.sin(x)
sinx
```

```python
pp.plot(x, sinx)
```

```python
cosx = np.cos(x)
cosx
```

```python
pp.plot(x, cosx)
```

```python
pp.plot(x, sinx, 'x')
pp.plot(x, cosx, 'o')
```

```python
# Mathematical operations
y = sinx * cosx
z = cosx **2 - sinx ** 2 # Differences of squares

# Plot results
pp.plot(x, y)
pp.plot(x, z)

# Add legend to the plot
pp.legend(['sinx * cosx', 'cosx **2 - sinx ** 2'])
```

```python
# Inner product of two vectors. That is the sum of the element-by-element products
np.dot(sinx, cosx)
```

```python
# Outer product, which builds every possible combination of the elements from the two vectors.
# The result is a matrix.
outer_product = np.outer(sinx, cosx)
outer_product
```

**ATTENTION**

Be very careful when performing mathematical operations with numpy arrays.  Python will try to 'guess' how to perform non-standard operations in the best way, and the result may not be what you would expect

```python
# Adding a number to a matrix or a vector
v = np.linspace(0, 10, 5)
v

# Adding a number to a vector or a matrix adds that number to every single element
v + 1
```

```python
# Adding a vector to a matrix
vv = np.outer(v, v)
vv

# In this case, Python will add the vector to each row
vv + v
```
