```python
import requests
from bs4 import BeautifulSoup
```

```python
# Make a request to http://courses.sci.pitt.edu/
# and get the HTML contents of the webpage
page = requests.get('http://courses.sci.pitt.edu/')
print(page.content)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
courses = soup.find_all('div', class_='course')
```

```python
print(len(courses))
```

```python
course = courses[0]
title = course.find('h3').get_text()
print(title)
```

```python
sections = course.find_all('div', class_='row section')[0].children
for section in sections:
    class_number = section.get_text()
    print(class_number)
```

```python
page = requests.get('http://courses.sci.pitt.edu/')
soup = BeautifulSoup(page.content, 'html.parser')
courses = soup.find_all('div', class_='course')
for course in courses:
    title = course.find('h3').get_text().strip()
    print(title)
    sections = course.find_all('div', class_='row section')[0].children
    for section in sections:
        section_info = section.get_text()
        print('\t' + section_info) 
    
```
