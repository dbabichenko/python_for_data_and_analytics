```python
import math
df['weight_transformed_sqrt'] = df['weight'].apply(lambda x: math.sqrt(x))
```

```python
sns.histplot(data=df['weight_transformed_sqrt'], kde=True)
```

```python
df['weight_transformed_log'] = df['weight'].apply(lambda x: math.log(x+1))
```

```python
sns.histplot(data=df['weight_transformed_log'], kde=True)
```

```python
df['weight_scaled'] = prep.scale(df['weight_transformed_log'])
```

```python
sns.histplot(data=df['weight_scaled'], kde=True)
```
