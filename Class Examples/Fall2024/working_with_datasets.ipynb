### Concat

**Pandas Cheat sheet**
https://pandas.pydata.org/Pandas_Cheat_Sheet.pdf

**Concat by rows**

**Assumptions: **
* Two or more dataframes
* Each dataframe has the same number of columns
* Columns have identical headings
* Columns in all dataframes appear in the same order

```python
df.head()
```

```python
df1 = df.iloc[:3, :5]
df1.head()
```

```python
df2 = df.iloc[4:6, :5]
df2.head()
```

```python
df3 = pd.concat([df1, df2], axis=0)
df3
```

### Concat by columns

**Concat by column**

**Assumptions: **
* Two or more dataframes
* Each dataframe has the same number of rows
* Rows have identical indexes
* Rows in all dataframes appear in the same order

```python
df.head()
```

```python
df1 = df.iloc[:3, [0, 1, 2]]
df1
```

```python
df2 = df.iloc[:3, [3, 4, 5]]
df2
```

```python
df3 = pd.concat([df1, df2], axis=1)
df3
```

### Merging DataFrame(s)

```python
df.head()
```

```python
df1 = df.iloc[:3, :3]
df1
```

```python
df2 = df.iloc[1:4, [0, 4, 5, 6]]
df2
```

```python
## Let's compare the two dataframes
df1.head()
```

```python
df2.head()
```

In Pandas, instead of using SQL keyword "JOIN", we use a method called "merge()"

```python
# Inner merge
df3 = pd.merge(df1, df2, how='inner', left_on='id', right_on='id')
df3
```

```python
# Left merge
df4 = pd.merge(df1, df2, how='left', left_on='id', right_on='id')
df4
```

```python
df4.isnull().sum()
```

```python
# Right merge
df5 = pd.merge(df1, df2, how='right', left_on='id', right_on='id')
df5
```

```python
df5 = pd.merge(df2, df1, how='left', left_on='id', right_on='id')
df5
```

```python
# Full join (outer join)
df6 = pd.merge(df1, df2, how='outer', left_on='id', right_on='id')
df6
```

### Querying

```python
df.head()
```

```python
df['id']
```

```python
df[['id', 'age', 'weight']]
```

```python
df.head()
```

```python
q = "chol > 350"
df.query(q)
```

```python
q = "chol > 300 and age <= 60 and gender == 'female'"
df.query(q)
```

```python
q = "(location == 'Buckingham' and frame == 'medium') or (gender == 'male' and weight > 200)"
df.query(q)
```

```python
df.query("age > 80")[['id', 'age', 'weight']]
```

```python
df.query("age > 80")[['id', 'age', 'weight']].query("weight >= 200")
```

```python
df.query("age > 80")[['age', 'weight']].plot.scatter(x='age', y='weight')
```
