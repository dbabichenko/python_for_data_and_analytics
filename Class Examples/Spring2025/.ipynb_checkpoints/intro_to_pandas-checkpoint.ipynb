## Combining Datasets
1. concat
2. merge

**Dire Warning: You should only use concat when rows or columns in both dataframes align!**

### Concat function 
#### Concat by row

```python
df.head()
```

```python
# Get a subset of the original dataframe
df1 = df.loc[0:2][['id', 'location', 'age', 'gender', 'height']]
df1.head()
```

```python
# Get a subset of the original dataframe
df2 = df.loc[3:6][['id', 'location', 'age', 'gender', 'height']]
df2.head()
```

```python
# Concat by rows 
df3 = pd.concat([df1, df2], axis=0)
df3
```

#### Concat by columns

```python
# Get a subset of the original dataframe
df4 = df.loc[0:3][['id', 'location']]
df4.head()
```

```python
# Get a subset of the original dataframe
df5 = df.loc[0:3][['age', 'gender', 'frame']]
df5.head()
```

```python
# Concat by columns
df6 = pd.concat([df4, df5], axis=1)
df6
```

#### What happens when rows or columns do not align?!

```python
df4.head()
```

```python
df1.head()
```

```python
df7 = pd.concat([df1, df4], axis=0)
df7
```

```python
df8 = pd.concat([df1, df4], axis=1)
df8
```

### Merging dataframes

```python
df1 = df.loc[0:2][['id', 'age', 'gender']]
df1
```

```python
df2 = df.loc[1:3][['id', 'chol', 'weight', 'frame']]
df2
```

```python
# Inner merge
df_inner = pd.merge(df1, df2, how='inner', left_on='id', right_on='id')
df_inner
```

```python
# Left merge
df_left = pd.merge(df1, df2, how='left', left_on='id', right_on='id')
df_left
```

```python
# Right merge
df_right = pd.merge(df1, df2, how='right', left_on='id', right_on='id')
df_right
```

```python
# Full (outer) merge
df_full = pd.merge(df1, df2, how='outer', left_on='id', right_on='id')
df_full
```

### Querying Pandas

```python
df.head()
```

```python
q = "age >= 30"  # A query is just a boolean expression
df.query(q)
```

```python
df.query("age >= 80")[['id', 'chol', 'age', 'hdl']]
```

```python
## Sorting
df.query("age >= 80")[['id', 'chol', 'age', 'hdl']].sort_values('chol', ascending=True)
```

```python
## Getting max or min value
# Ex: Find a patient who is older than 80 and has the smallest chol in the group
df.query("age > 80")[['id', 'chol', 'age', 'hdl']].sort_values('chol', ascending=True).head(1)
```

```python
## Get female patients who live in Buckingham and are between the ages of 20 and 30
q = "age >= 20 and age <= 30 and location == 'Buckingham'"
df.query(q)
```

```python
## Get female patients who live in Buckingham 
## and are between the ages of 20 and 30
## and have chol of 160 or less
q = "age >= 20 and age <= 30 and location == 'Buckingham' and chol <= 160"
df.query(q)
```

```python
## Find all patients who live in towns that begin with a letter 'B'
df.query('location.str.startswith("B")')
```

### Grouping and Aggregation

### Aggregation functions:
* min
* max
* mean
* count
* sum

```python
### Find out how many patients are in each location
df.value_counts('location')
```

```python
df.groupby('location').agg({'age' : 'mean', 'weight' : 'max', 'chol' : 'min'})
```

```python
# Find all patients that are male
# Group the results by location
# Aggregate: age --> mean, weight --> max, chol --> min
df.query('gender == "male"').groupby('location').agg({'age' : 'mean', 'weight' : 'max', 'chol' : 'min'})
```

```python
# Find all patients that are male
# Group the results by location
# Aggregate: age --> mean, weight --> max, chol --> min
# Sort by chol in descending order
df.query('gender == "male"').groupby('location').agg({'age' : 'mean', 'weight' : 'max', 'chol' : 'min'}).sort_values('chol', ascending=False)
```

### Iterating through dataframe

```python
# iterating through a list
l = [1, 2, 'x', 'Hello', 3.5]

for item in l:
    print(item)
```

```python
# iterating through a dictionary
d = {
    'a' : 5,
    'b' : 10,
    'c' : 2343
}

#d.items()

for key, value in d.items():
    print(key, " --> ", value)

```

```python
# iterating through dataframes
for index, row in df.iterrows():
    print(index, ' -->', row['id'], ': ', row['gender'])
```
