## Combining Datasets
1. concat
2. merge

**Dire Warning: You should only use concat when rows or columns in both dataframes align!**

### Concat function 
#### Concat by row

```python
df.head()
```

```python
# Get a subset of the original dataframe
df1 = df.loc[0:2][['id', 'location', 'age', 'gender', 'height']]
df1.head()
```

```python
# Get a subset of the original dataframe
df2 = df.loc[3:6][['id', 'location', 'age', 'gender', 'height']]
df2.head()
```

```python
# Concat by rows 
df3 = pd.concat([df1, df2], axis=0)
df3
```

#### Concat by columns

```python
# Get a subset of the original dataframe
df4 = df.loc[0:3][['id', 'location']]
df4.head()
```

```python
# Get a subset of the original dataframe
df5 = df.loc[0:3][['age', 'gender', 'frame']]
df5.head()
```

```python
# Concat by columns
df6 = pd.concat([df4, df5], axis=1)
df6
```

### Merging dataframes

```python
df1 = df.loc[0:2][['id', 'age', 'gender']]
df1
```

```python
df2 = df.loc[1:3][['id', 'chol', 'weight', 'frame']]
df2
```

```python
# Inner merge
df_inner = pd.merge(df1, df2, how='inner', left_on='id', right_on='id')
df_inner
```

```python
# Left merge
df_left = pd.merge(df1, df2, how='left', left_on='id', right_on='id')
df_left
```

```python
# Right merge
df_right = pd.merge(df1, df2, how='right', left_on='id', right_on='id')
df_right
```

```python
# Full (outer) merge
df_full = pd.merge(df1, df2, how='outer', left_on='id', right_on='id')
df_full
```
