```python
stocks = pd.read_csv('stocks.csv')
stocks.head()
```

```python
stocks.select_dtypes(include='number').corr()
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(stocks.select_dtypes(include='number'), # Make a scatter matrix of all columns
               figsize=(30, 30), # Set plot size
               diagonal='kde')  # Show distribution estimates on diagonal
```

```python
stocks.plot.scatter(x='open', y='close')
```

```python
stocks.plot.scatter(x='open', y='volume')
```

```python
sn.scatterplot(data=stocks, x='open', y='close')
```

```python
import itertools
```

```python

num_columns = list(stocks.select_dtypes(include='number').columns)
combs = list(itertools.combinations(num_columns, 2))
for c in combs:
    sn.scatterplot(data=stocks, x = c[0], y=c[1])
    plt.show()
```

```python
sn.scatterplot(data=stocks, x='open', y='close', s=stocks['volume'] * 0.0005, c=stocks['high'])
```
