```python
# Required Python Machine learning Packages
import pandas as pd
import numpy as np

# To split the dataset into train and test datasets
from sklearn.model_selection import train_test_split

# To model the Gaussian Navie Bayes classifier
from sklearn.naive_bayes import GaussianNB

# To calculate the accuracy score of the model
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

import matplotlib.pyplot as plt
import seaborn as sn
%matplotlib inline
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
```

```python
def label_encode(df, columns):
    for col in columns:
        le = LabelEncoder()
        col_values_unique = list(df[col].unique())
        le_fitted = le.fit(col_values_unique)
 
        col_values = list(df[col].values)
        #le.classes_
        col_values_transformed = le.transform(col_values)
        new_col_name = col + '_le'
        df[new_col_name] = col_values_transformed
        
```

```python
df = pd.read_csv('pet_adoption_dataset.csv')
df.head()
```

```python
df[['animal_type', 'animal_type_le']].groupby(['animal_type', 'animal_type_le']).agg({'animal_type_le': 'count'})
```

```python
X = df[['animal_type_le', 'age_months', 'weight_kg']]	
y = df[['adopted']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 10)
```

```python
nb = GaussianNB()
nb.fit(X_train, y_train)
y_pred = nb.predict(X_test)
y_pred
```

```python
print(metrics.accuracy_score(y_test, y_pred))
```

```python
print(metrics.roc_auc_score(y_test, y_pred))
```
