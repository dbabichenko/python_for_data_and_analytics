```python
import requests

from bs4 import BeautifulSoup
```

```python
base_url = "http://www.pitt.edu/~dmb72/table.html"
page = requests.get(base_url)

soup = BeautifulSoup(page.content, 'html.parser')

```

```python
print(soup.prettify())
```

```python
columns = []
data = []

# Get all rows
rows = soup.find_all('tr')

# We know that the first row of the table contains column headers
# In the first row, find all elements <th>
col_headers = rows[0].find_all('th')

# Create a list of column headers
for col in col_headers:
    columns.append(col.get_text())

#print(columns)

# Iterate through rows 1 through all (skip row 0)
for row in rows[1:]:
    one_row = []
    cells = list(row.find_all('td'))
    
    # The following code block is a bit of a hack and not the proper way of doing things
    '''
    for i in range(0, len(cells)):
        cells[i] = str(cells[i]).replace('<td>', '').replace('</td>', '')
    print(cells)
    break
    
    '''
    for cell in cells:
        # Adds individual cell values to a row
        one_row.append(cell.get_text())
    data.append(one_row)
    
print(data)
```
